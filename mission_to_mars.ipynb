```python
from splinter import Browser 
from bs4 import BeautifulSoup
import pandas as pd 
```

NASA Mars News

```python
executable_path={'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://mars.nasa.gov/news/'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
results = soup.find_all('div', class_='list_text')
print(results)
```

```python
for result in results:
    try:
        for div in soup.find_all('div', class_='content_title'):
            news_title = div.a.text
            #print(news_title)
        for text_body in soup.find_all('div', class_='article_teaser_body'):
            news_p = text_body.text
            
            #print(news_p.text)
            if (news_title or news_p):
                print('--------------------')
                print(news_title)
                print(news_p)                        

    except AttributeError as e:
        print(e)
        
        
        
browser.click_link_by_partial_text('Scaping Complete')
```

Mars Weather

```python
executable_path={'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://twitter.com/marswxreport'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
results = soup.find_all('div', class_='js-tweet-text-container')
print(results)
```

```python
for result in results:
    try:
        mars_weather = result.find('p', class_='TweetTextSize TweetTextSize--normal js-tweet-text tweet-text').text
        
        print('mars_weather = ' + mars_weather)
        print('-------------------------')
    except AttributeError as e:
        print(e)
        
        
```
